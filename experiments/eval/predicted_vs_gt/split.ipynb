```python
import pandas as pd
from pathlib import Path

in_path = Path("Claude_eval_with_gt.csv")
df = pd.read_csv(in_path)

# Robust match: treat NaN as empty; case-insensitive contains "hourly"
mask_multi = df["query_id"].astype(str).str.contains("hourly", case=False, na=False)

df_multi_step = df.loc[mask_multi].copy()
df_point = df.loc[~mask_multi].copy()

out_multi = in_path.with_name("Claude_multi_step.csv")
out_point = in_path.with_name("Claude_point_forecast.csv")

df_multi_step.to_csv(out_multi, index=False)
df_point.to_csv(out_point, index=False)

{
    "input_rows": len(df),
    "multi_step_rows": len(df_multi_step),
    "point_forecast_rows": len(df_point),
    "multi_step_file": str(out_multi),
    "point_forecast_file": str(out_point),
    "sample_multi_query_ids": df_multi_step["query_id"].head(5).tolist() if "query_id" in df_multi_step.columns else None,
    "sample_point_query_ids": df_point["query_id"].head(5).tolist() if "query_id" in df_point.columns else None,
}
```
