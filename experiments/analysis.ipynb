# Cost and Latency

```python
import pandas as pd

FILES = {
    "OpenAI": "OpenAI/OpenAI_eval_with_api.csv",
    "Claude": "Claude/Claude_eval_with_api.csv",
    "Gemini": "Gemini/Gemini_eval_with_api.csv",
    "DeepSeek": "DeepSeek/DeepSeek_eval_with_api.csv",
}

STAGE_ORDER = ["reproducibility", "horizon", "pattern", "stats", "summary"]

out = []
for model, path in FILES.items():
    df = pd.read_csv(path)

    # ensure numeric
    df["cost_usd"] = pd.to_numeric(df["cost_usd"], errors="coerce")
    df["latency_sec"] = pd.to_numeric(df["latency_sec"], errors="coerce")

    for stage in STAGE_ORDER:
        sub = df[df["stage"] == stage]
        out.append({
            "model": model,
            "stage": stage,
            "n": len(sub),
            "cost_mean": sub["cost_usd"].mean(),
            "lat_mean": sub["latency_sec"].mean(),
            "cost_std": sub["cost_usd"].std(),
            "lat_std": sub["latency_sec"].std(),
        })

res = pd.DataFrame(out)
print(res)
```

```python
res.to_csv("cost_latency_by_stage.csv", index=False)
print("Saved to: cost_latency_by_stage.csv")
```

# Merge predicted with ground truth

```python
import pandas as pd

DEEPSEEK_PATH = "DeepSeek/DeepSeek_eval_with_api.csv"
QUERIES_PATH = "queries_from_json.csv"
OUTPUT_PATH = "DeepSeek_eval_with_gt.csv"

deepseek = pd.read_csv(DEEPSEEK_PATH)
queries = pd.read_csv(QUERIES_PATH)

print("DeepSeek columns:", list(deepseek.columns))
print("Queries columns:", list(queries.columns))

# Validate required columns
if "query_id" not in deepseek.columns:
    raise ValueError("DeepSeek file must contain 'query_id'.")

if "id" not in queries.columns:
    raise ValueError("Queries file must contain 'id'.")

if "ground_truth" not in queries.columns:
    raise ValueError("Queries file must contain 'ground_truth'.")

# Ensure uniqueness in queries
if queries["id"].duplicated().any():
    raise ValueError("Duplicate ids found in queries file.")

# Prepare ground truth mapping
gt = queries[["id", "ground_truth"]].rename(columns={"id": "query_id"})

# Merge
merged = deepseek.merge(
    gt,
    on="query_id",
    how="left",
    validate="many_to_one"
)

print("Rows in DeepSeek:", len(deepseek))
print("Rows after merge:", len(merged))
print("Missing ground_truth values:", merged["ground_truth"].isna().sum())

# Save
merged.to_csv(OUTPUT_PATH, index=False)

print("Saved to:", OUTPUT_PATH)

merged.head()
```
