```python
import pandas as pd

def load_data(file_path: str) -> pd.DataFrame:
    """Load data from a CSV file."""
    return pd.read_csv(file_path)

df=load_data("data/processed_data.csv")
```

```python
# split dataframe by specific regions into a dict of DataFrames (always create all 5)
regions = ['NSW1', 'TAS1', 'QLD1', 'VIC1', 'SA1']
regions_dfs = {}
for r in regions:
    if r in df['REGION'].unique():
        regions_dfs[r] = df[df['REGION'] == r].reset_index(drop=True)
    else:
        # create an empty DataFrame with same columns if region not present
        regions_dfs[r] = df.head(0).copy()
regions_dfs
```

```python
# find the newest (latest) settlement datetime and show corresponding rows
settlement_dt = pd.to_datetime(df['SETTLEMENTDATE'])
newest_dt = settlement_dt.max()
newest_rows = df[settlement_dt == newest_dt].reset_index(drop=True)

print("Newest SETTLEMENTDATE:", newest_dt)
newest_rows
```

```python
from scipy import stats

# Relationship and regional breakdowns for demand and price

# use existing settlement_dt for consistent datetimes
df_stats = df.copy()
df_stats['SETTLEMENTDATE'] = settlement_dt

# add time helpers
df_stats['YEAR'] = df_stats['SETTLEMENTDATE'].dt.year
df_stats['MONTH'] = df_stats['SETTLEMENTDATE'].dt.month
df_stats['DAY'] = df_stats['SETTLEMENTDATE'].dt.date
df_stats['YM'] = df_stats['SETTLEMENTDATE'].dt.to_period('M')

# 1) Relationship between Demand and Price
demand_price_corr_overall = df_stats['TOTALDEMAND'].corr(df_stats['RRP'])
demand_price_corr_by_region = df_stats.groupby('REGION').apply(
    lambda g: g['TOTALDEMAND'].corr(g['RRP'])
).rename('pearson_corr')

# 2) Price by region: aggregated by year, by month (year-month), and by calendar day
price_by_region_year = df_stats.groupby(['REGION', 'YEAR'])['RRP'].agg(['mean','median','std','min','max','count'])
price_by_region_month = df_stats.groupby(['REGION', 'YM'])['RRP'].agg(['mean','median','std','min','max','count'])
price_by_region_day = df_stats.groupby(['REGION', 'DAY'])['RRP'].agg(['mean','median','std','count'])

# 3) Regional differences (summary and a simple ANOVA on TOTALDEMAND across regions)
regional_means = df_stats.groupby('REGION')[['TOTALDEMAND','RRP']].mean()
regional_medians = df_stats.groupby('REGION')[['TOTALDEMAND','RRP']].median()
regional_std = df_stats.groupby('REGION')[['TOTALDEMAND','RRP']].std()

# ANOVA (warning: may be slow on full dataset)
try:
    groups = [g['TOTALDEMAND'].values for _, g in df_stats.groupby('REGION')]
    anova_f, anova_p = stats.f_oneway(*groups)
except Exception as e:
    anova_f, anova_p = (None, None)

# 4) Average monthly demand (national)
average_monthly_demand = df_stats.groupby('YM')['TOTALDEMAND'].mean().rename('avg_monthly_demand')

# 5) Regional demand comparison:
#    - monthly average demand per region (wide format)
monthly_avg_by_region = df_stats.groupby(['YM','REGION'])['TOTALDEMAND'].mean().unstack('REGION').fillna(0)
#    - percent share of national monthly demand by region
monthly_total = monthly_avg_by_region.sum(axis=1)
monthly_pct_by_region = monthly_avg_by_region.div(monthly_total, axis=0).fillna(0)

# Package outputs for inspection
analysis = {
    'demand_price_corr_overall': demand_price_corr_overall,
    'demand_price_corr_by_region': demand_price_corr_by_region,
    'price_by_region_year': price_by_region_year,
    'price_by_region_month': price_by_region_month,
    'price_by_region_day': price_by_region_day,
    'regional_means': regional_means,
    'regional_medians': regional_medians,
    'regional_std': regional_std,
    'anova_f': anova_f,
    'anova_p': anova_p,
    'average_monthly_demand': average_monthly_demand,
    'monthly_avg_by_region': monthly_avg_by_region,
    'monthly_pct_by_region': monthly_pct_by_region
}

print("analysis keys:", list(analysis.keys()))
analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Safety sampling for scatter/box plots to keep plotting fast
_sample_n = min(20000, len(df_stats))

fig, axes = plt.subplots(3, 2, figsize=(16, 14))

# 1) National average monthly demand (time series)
ax = axes[0, 0]
average_monthly_demand.plot(ax=ax, marker='o', linewidth=1)
ax.set_title("National Average Monthly Demand")
ax.set_ylabel("Avg TOTALDEMAND")
ax.set_xlabel("Month")

# 2) Stacked area: percent share of monthly demand by region
ax = axes[0, 1]
monthly_pct_by_region.plot.area(ax=ax, alpha=0.85)
ax.set_title("Monthly Percent Share of National Demand by Region (stacked)")
ax.set_ylabel("Percent share")
ax.set_xlabel("Month")
ax.legend(title="Region", bbox_to_anchor=(1.02, 1), loc="upper left")

# 3) Regional mean total demand (bar)
ax = axes[1, 0]
regional_means['TOTALDEMAND'].sort_values(ascending=False).plot.bar(ax=ax, color=sns.color_palette("muted"))
ax.set_title("Mean TOTALDEMAND by Region")
ax.set_ylabel("Mean TOTALDEMAND")
ax.set_xlabel("Region")

# 4) Boxplot of RRP by region (sampled)
ax = axes[1, 1]
sns.boxplot(x="REGION", y="RRP", data=df_stats.sample(n=_sample_n, random_state=42), ax=ax)
ax.set_title(f"Distribution of RRP by Region (sample {_sample_n:,})")
ax.set_ylabel("RRP (price)")
ax.set_xlabel("Region")

# 5) Scatter: TOTALDEMAND vs RRP with a regression line (sampled)
ax = axes[2, 0]
_sample = df_stats.sample(n=_sample_n, random_state=1)
sns.scatterplot(x="TOTALDEMAND", y="RRP", hue="REGION", data=_sample, ax=ax, s=15, alpha=0.5, palette="tab10", legend="brief")
sns.regplot(x="TOTALDEMAND", y="RRP", data=_sample, scatter=False, ax=ax, color="black", line_kws={"linewidth":1.5})
ax.set_title("TOTALDEMAND vs RRP (sampled) with overall regression")
ax.set_xlabel("TOTALDEMAND")
ax.set_ylabel("RRP")

# 6) Heatmap: correlation of monthly average demand between regions
ax = axes[2, 1]
corr_mat = monthly_avg_by_region.corr()
sns.heatmap(corr_mat, annot=True, fmt=".2f", cmap="vlag", center=0, ax=ax)
ax.set_title("Correlation of Monthly Average Demand Between Regions")

plt.tight_layout()
plt.show()

# Print a quick summary and the newest rows (most recent timestamp)
print("Overall demand-price Pearson correlation:", float(analysis['demand_price_corr_overall']))
print("\nPearson correlation by region:\n", analysis['demand_price_corr_by_region'])
print("\nNewest settlement rows:")
print(newest_rows)
```
